## **Course Description**  
This course introduces fundamental concepts and techniques in unsupervised machine learning, covering feature engineering, dimensionality reduction, clustering, and model evaluation. Students will learn how to preprocess data, reduce dimensions using PCA, apply clustering algorithms like K-Means, and evaluate models using metrics like the Silhouette Score. The course also covers cross-validation and hyperparameter tuning for unsupervised learning tasks.  

---

## **Course Objectives**  
By the end of this course, students will be able to:  
1. Understand the role of unsupervised learning in machine learning.  
2. Perform feature engineering and selection for unsupervised tasks.  
3. Apply dimensionality reduction techniques like PCA.  
4. Implement clustering algorithms such as K-Means.  
5. Evaluate unsupervised models using appropriate metrics.  
6. Apply cross-validation and hyperparameter tuning in unsupervised learning.
